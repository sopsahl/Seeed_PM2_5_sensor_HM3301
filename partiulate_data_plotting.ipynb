```python
import numpy as np
import matplotlib.pyplot as plt
import mysql.connector as sq
from datetime import datetime
from matplotlib.dates import DateFormatter
```

```python
db_config = {
    'host': '18.25.16.44',
    'port': 3307,
    'user': 'lab',
    'password': 'qn&ngpwdDB1',
    'database': 'qnndb'
}

# Open the .sql file for reading

# ## Connect to the MySQL database
conn = sq.connect(**db_config)
cursor = conn.cursor()

# # Execute the SQL query
cursor.execute("SELECT * FROM `particulate_aja_room` ORDER BY 'id' LIMIT 1188 OFFSET 5378")
# 5378,6565
# # Fetch the data and store it in a list
result = np.array(cursor.fetchall())

# # Commit the changes and close the connection
conn.commit()
conn.close()
```

```python
date_format = DateFormatter('%H')
```

```python
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.plot(result[:, 1], result[:, 2])
plt.xlabel("Time")
plt.ylabel('Humidity (%)')
plt.title("Humidity over 10/30 and 10/31 in the aja room")
plt.show()
```

```python
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.plot(result[:, 1], result[:, 3])
plt.xlabel("Time")
plt.ylabel('Temperature (deg C)')
plt.title("Temperature over 10/30 and 10/31 in the aja room")
plt.show()
```

```python
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.plot(result[:, 1], result[:, 6], label = "< 10 microns")
plt.plot(result[:, 1], result[:, 5],  label = "< 2.5 microns")
plt.plot(result[:, 1], result[:, 4], label = "< 1 micron")
plt.legend()
plt.xlabel("Time")
plt.ylabel('Relative Count')
plt.title("Standard Particulate Matter over 10/30 and 10/31 in the aja room")
plt.show()
```

```python
plt.gca().xaxis.set_major_formatter(date_format)
plt.gcf().autofmt_xdate()
plt.plot(result[:, 1], result[:, 9], label = "< 10 microns")
plt.plot(result[:, 1], result[:, 8],  label = "< 2.5 microns")
plt.plot(result[:, 1], result[:, 7], label = "< 1 micron")
plt.legend()
plt.xlabel("Time")
plt.ylabel('Relative Count')
plt.title("Atmospheric Particulate Matter over 10/30 and 10/31 in the aja room")
plt.show()
```
